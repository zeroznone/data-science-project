### Adding columns

```python
train['ones']=1
```

```python
t=train.groupby(['Patient-Uid','Date','Incident']).sum()
```

```python
final_train=t.unstack()
```

```python
final_train.fillna(0,inplace=True)
```

```python
final_train
```

```python
final_train.columns=final_train.columns.droplevel()
```

```python
X=final_train.drop(['TARGET DRUG'],axis=1)
y=final_train['TARGET DRUG']
```

```python
X.columns
```

## TRain split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

# Ridge Classifier

```python
from sklearn.linear_model import RidgeClassifier
clf = RidgeClassifier().fit(X_train, y_train)
```

```python
pred=clf.predict(X_test)
```

## F1 score

```python
pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, pred, average='weighted')
```
